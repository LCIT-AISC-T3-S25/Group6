```python
import pandas as pd
import re


file_path = '/content/drive/MyDrive/Data.csv'
df = pd.read_csv(file_path, encoding='utf-8', engine='python')


emoji_pattern = r":\)|:D|;\)|:P"


emoji_tweet_count = df['Data'].astype(str).apply(lambda x: bool(re.search(emoji_pattern, x))).sum()

print("Number of tweets that contain at least one of the emojis :), :D, ;), :P:", emoji_tweet_count)
```
