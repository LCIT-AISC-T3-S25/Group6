# How many records make a word that starts with an alphabet and 
is not a URL?

```python
import pandas as pd
import re
```

```python
df = pd.read_csv("Data.csv")
df.head()
```

```python
# Pattern to detect URLs
url_pattern = r'http[s]?://\S+'

# Pattern to find words that start with an alphabet
word_pattern = r'\b[a-zA-Z]\w*'
```

```python
def is_valid_text(text):
    has_url = re.search(url_pattern, text)
    has_word = re.search(word_pattern, text)
    return has_word and not has_url
```

```python
df['is_valid'] = df['Data'].apply(is_valid_text)
```

```python
valid_count = df['is_valid'].sum()
print("Number of valid records:", valid_count)
```

# Perform stemming and lemmatization. Compare the change in word frequencies from stemming and lemmatization. Which are the top 10 words after stemming/lemmatization


```python
#Importing Required Libraries
import pandas as pd
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk import pos_tag
from collections import Counter
```

```python
# Downloading Required NLTK Resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')
```

```python
#Initializing Stemmer and Lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
#POS Mapping Function for Lemmatization
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default
```

```python
#Define Tokenization + Transformation Functions
def stem_text(text):
    tokens = word_tokenize(str(text).lower())
    return [stemmer.stem(word) for word in tokens]

def lemmatize_text(text):
    tokens = word_tokenize(str(text).lower())
    pos_tags = pos_tag(tokens)
    return [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]
```

```python
#Applying Stemming and Lemmatization
df['Stemmed_Tokens'] = df['Data'].apply(stem_text)
df['Lemmatized_Tokens'] = df['Data'].apply(lemmatize_text)
```

```python
#Flatten Word Lists for Frequency Count
all_stemmed_words = [word for row in df['Stemmed_Tokens'] for word in row]
all_lemmatized_words = [word for row in df['Lemmatized_Tokens'] for word in row]
```

```python
#Calculate Word Frequencies
stemmed_freq = Counter(all_stemmed_words)
lemmatized_freq = Counter(all_lemmatized_words)
```

```python
#Displaying Top 10 Most Frequent Words
print("🔹 Top 10 Words After Stemming:")
for word, count in stemmed_freq.most_common(10):
    print(f"{word}: {count}")

print("\n🔹 Top 10 Words After Lemmatization:")
for word, count in lemmatized_freq.most_common(10):
    print(f"{word}: {count}")
```
