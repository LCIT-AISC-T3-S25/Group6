Q1. How many records have a date that is expressed without using alphabets?

```python
import pandas as pd
import re
data = pd.read_csv("Data.csv")
data.head()

# Comprehensive regex pattern for dates
date_pattern = r"""
    # Match dates with slashes, dashes or dots
    (?:
        # MM/DD/YY or MM/DD/YYYY (American format)
        (?:0?[1-9]|1[0-2])[/\-\.](?:0?[1-9]|[12][0-9]|3[01])[/\-\.](?:19|20)?\d\d
        |
        # DD/MM/YY or DD/MM/YYYY (European format)
        (?:0?[1-9]|[12][0-9]|3[01])[/\-\.](?:0?[1-9]|1[0-2])[/\-\.](?:19|20)?\d\d
        |
        # YYYY/MM/DD (ISO format)
        (?:19|20)\d\d[/\-\.](?:0?[1-9]|1[0-2])[/\-\.](?:0?[1-9]|[12][0-9]|3[01])
        |
        # YYYYMMDD (Compact ISO format)
        (?:19|20)\d\d(?:0[1-9]|1[0-2])(?:0[1-9]|[12][0-9]|3[01])
    )
"""

# Compiling the regex pattern first
compiled_pattern = re.compile(date_pattern, re.VERBOSE)

# Using a lambda function to apply the compiled pattern
date_mask = data['Data'].apply(lambda x: bool(compiled_pattern.search(str(x))) if pd.notna(x) else False)

# Counting the instances that match
date_count = date_mask.sum()
if date_count > 0:
    # Displaying all matching records
    date_records = data[date_mask]
    print("Records with dates (no alphabets):")
    display(date_records)
    print("\n")
print("Total number of records with dates expressed without alphabets:", date_count)
```

There is no records with just numeric dates

Q2. Perform lemmatization using NLTK lemmatizer. Count the number of

```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download required NLTK data files (only needed once)
# Added the download for 'punkt_tab' which is required by word_tokenize internally
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab') # Download the punkt_tab resource

# Load the CSV file
df = pd.read_csv("Data.csv")

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Combine all text data into a single string
all_text = ' '.join(df['Data'].dropna().astype(str))

# Tokenize the text
tokens = word_tokenize(all_text)

# Count unique tokens before lemmatization
unique_before = set(tokens)

# Lemmatize each token
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

# Count unique tokens after lemmatization
unique_after = set(lemmatized_tokens)

# Print the counts
print("Unique tokens before lemmatization:", len(unique_before))
print("Unique tokens after lemmatization:", len(unique_after))
```

Q3. Compare the change in word frequencies from stemming and
lemmatization. Which are the top 10 words after stemming/lemmatization

```python
#Importing Required Libraries
import pandas as pd
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk import pos_tag
from collections import Counter
# Downloading Required NLTK Resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('omw-1.4')

#Initializing Stemmer and Lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
#POS Mapping Function for Lemmatization
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # default
#Define Tokenization + Transformation Functions
def stem_text(text):
    tokens = word_tokenize(str(text).lower())
    return [stemmer.stem(word) for word in tokens]

def lemmatize_text(text):
    tokens = word_tokenize(str(text).lower())
    pos_tags = pos_tag(tokens)
    return [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]
#Applying Stemming and Lemmatization
df['Stemmed_Tokens'] = df['Data'].apply(stem_text)
df['Lemmatized_Tokens'] = df['Data'].apply(lemmatize_text)
#Flatten Word Lists for Frequency Count
all_stemmed_words = [word for row in df['Stemmed_Tokens'] for word in row]
all_lemmatized_words = [word for row in df['Lemmatized_Tokens'] for word in row]
#Calculate Word Frequencies
stemmed_freq = Counter(all_stemmed_words)
lemmatized_freq = Counter(all_lemmatized_words)
#Displaying Top 10 Most Frequent Words
print("🔹 Top 10 Words After Stemming:")
for word, count in stemmed_freq.most_common(10):
    print(f"{word}: {count}")

print("\n🔹 Top 10 Words After Lemmatization:")
for word, count in lemmatized_freq.most_common(10):
    print(f"{word}: {count}")
```

Q4. What is the change in word frequencies if normalization is done after
stop word removal

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk import pos_tag
from nltk.stem import PorterStemmer, WordNetLemmatizer
from collections import Counter
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

# Text Processing Class
class TextNormalizer:
    def __init__(self):
        self.stop_words = set(stopwords.words('english'))
        self.stemmer = PorterStemmer()
        self.lemmatizer = WordNetLemmatizer()

    def tokenize(self, text):
        return word_tokenize(str(text).lower())

    def remove_stopwords(self, tokens):
        return [t for t in tokens if t.isalnum() and t not in self.stop_words]

    def stem(self, tokens):
        return [self.stemmer.stem(t) for t in tokens]

    def lemmatize(self, tokens):
        tagged = pos_tag(tokens)
        return [self.lemmatizer.lemmatize(word, self.get_wordnet_pos(tag)) for word, tag in tagged]

    def get_wordnet_pos(self, tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        return wordnet.NOUN  # default fallback

    def normalize(self, text, method='stem'):
        tokens = self.tokenize(text)
        tokens = self.remove_stopwords(tokens)
        if method == 'stem':
            return self.stem(tokens)
        elif method == 'lemma':
            return self.lemmatize(tokens)
        else:
            raise ValueError("Method must be either 'stem' or 'lemma'.")
df = pd.read_csv("Data.csv")
normalizer = TextNormalizer()
df['Stemmed_NoStop'] = df['Data'].apply(lambda x: normalizer.normalize(x, method='stem'))
df['Lemmatized_NoStop'] = df['Data'].apply(lambda x: normalizer.normalize(x, method='lemma'))
```
