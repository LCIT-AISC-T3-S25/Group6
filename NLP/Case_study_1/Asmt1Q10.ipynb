```python
import pandas as pd
import re

# Load the CSV file
df = pd.read_csv("Data.csv")

# Define a regular expression pattern to match URLs
url_pattern = r"http[s]?://\S+"

# Count the number of URLs in each tweet and sum them up
total_urls = df['Data'].str.count(url_pattern).sum()

print("Total number of URLs across all tweets:",total_urls)
```
