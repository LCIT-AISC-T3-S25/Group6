```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/Data.csv'
df = pd.read_csv(file_path, encoding='utf-8', engine='python')
```

```python
import re
hashtag_pattern = r"#\w+"
hashtag_count = df['Data'].astype(str).apply(lambda x: len(re.findall(hashtag_pattern, x))).sum()
print("Total number of hashtags across all tweets:", hashtag_count)
```
