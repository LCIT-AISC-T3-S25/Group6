```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download required NLTK data files (only needed once)
# Added the download for 'punkt_tab' which is required by word_tokenize internally
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab') # Download the punkt_tab resource

# Load the CSV file
df = pd.read_csv("Data.csv")

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Combine all text data into a single string
all_text = ' '.join(df['Data'].dropna().astype(str))

# Tokenize the text
tokens = word_tokenize(all_text)

# Count unique tokens before lemmatization
unique_before = set(tokens)

# Lemmatize each token
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

# Count unique tokens after lemmatization
unique_after = set(lemmatized_tokens)

# Print the counts
print("Unique tokens before lemmatization:", len(unique_before))
print("Unique tokens after lemmatization:", len(unique_after))
```
