```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r'C:\Users\Nitro\Downloads\Data.csv')
```

```python
# Combine all columns in each row into a single string
combined_rows = df.astype(str).agg(' '.join, axis=1)
```

```python
# Define regex pattern for decimal numbers
pattern = r'\d+\.\d+'
```

```python
# Count how many records contain a decimal number
decimal_count = combined_rows.str.contains(pattern).sum()

print(f"Number of records containing a decimal number: {decimal_count}")
```
