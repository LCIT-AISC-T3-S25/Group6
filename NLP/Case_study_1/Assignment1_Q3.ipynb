```python
import pandas as pd
import re
```

```python
df = pd.read_csv("Data.csv")
df.head()
```

```python
# Pattern to detect URLs
url_pattern = r'http[s]?://\S+'

# Pattern to find words that start with an alphabet
word_pattern = r'\b[a-zA-Z]\w*'
```

```python
def is_valid_text(text):
    has_url = re.search(url_pattern, text)
    has_word = re.search(word_pattern, text)
    return has_word and not has_url
```

```python
df['is_valid'] = df['Data'].apply(is_valid_text)
```

```python
valid_count = df['is_valid'].sum()
print("Number of valid records:", valid_count)
```
