```python
folder_path = "preprocessed_photos"
```

```python
import cv2
import os
import matplotlib.pyplot as plt
%matplotlib inline

# Dictionary to store images
images = {}

# Load images
for filename in os.listdir(folder_path):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):
        filepath = os.path.join(folder_path, filename)
        img = cv2.imread(filepath)
        if img is not None:
            images[filename] = img
        else:
            print(f"Failed to load {filename}")

print(f"Loaded {len(images)} images")
```

```python
# Gaussian Blur
def apply_gaussian_blur(img, kernel_size=(5, 5), sigma=0):
    return cv2.GaussianBlur(img, kernel_size, sigma)
```

```python
# Histogram Equalization
def apply_histogram_equalization(img):
    img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
    img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
    return cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
```

```python
# Intensity Thresholding
def apply_threshold(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return cv2.cvtColor(thresh, cv2.COLOR_GRAY2BGR)
```

```python
# Display function
def display_image(img, title="Image"):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6, 6))
    plt.imshow(img_rgb)
    plt.title(title)
    plt.axis("off")
    plt.show()

# Process all images
for filename, img in images.items():
    # Original
    display_image(img, title=f"{filename} - Original")
    
    # Gaussian Blur
    blurred = apply_gaussian_blur(img)
    display_image(blurred, title=f"{filename} - Gaussian Blur")
    
    # Histogram Equalization
    equalized = apply_histogram_equalization(img)
    display_image(equalized, title=f"{filename} - Histogram Equalization")
    
    # Intensity Thresholding
    thresh = apply_threshold(img)
    display_image(thresh, title=f"{filename} - Intensity Thresholding")
```

```python
output_dir = "Advanced _Processed_images"
os.makedirs(output_dir, exist_ok=True)

for filename, img in images.items():
    blurred = apply_gaussian_blur(img)
    equalized = apply_histogram_equalization(img)
    thresh = apply_threshold(img)
    
    # Save each version
    base_filename = os.path.splitext(filename)[0]
    cv2.imwrite(os.path.join(output_dir, f"{base_filename}_blurred.jpg"), blurred)
    cv2.imwrite(os.path.join(output_dir, f"{base_filename}_equalized.jpg"), equalized)
    cv2.imwrite(os.path.join(output_dir, f"{base_filename}_thresholded.jpg"), thresh)

print(f"Processed images saved to {output_dir}")
```

```python
import shutil
import os

# Zip the folder
shutil.make_archive("Advanced _Processed_images", "zip", "Advanced _Processed_images")


print("Zip file created at:", os.path.join(os.getcwd(), "Advanced _Processed_images.zip"))
```
